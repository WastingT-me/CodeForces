```python
def can_be_important_number(a):
    str_a = str(a)

    if str_a.startswith("10") and len(str_a) > 2:
        rest = str_a[2:]
        if not rest.startswith('0') and int(rest) >= 2:
            return "YES"

    return "NO"

def run_tests():
    test_cases = [
        (100, "NO"),
        (1010, "YES"),
        (101, "NO"),
        (105, "YES"),
        (2033, "NO"),
        (1019, "YES"),
        (1002, "NO")
    ]

    for i, (input_value, expected) in enumerate(test_cases):
        result = can_be_important_number(input_value)
        assert result == expected, f"Test case {i+1} failed: input({input_value}) => output({result}), expected({expected})"

if __name__ == "__main__":
    #run_tests()

    t = int(input())
    numbers = [int(input()) for _ in range(t)]
    for number in numbers:
        print(can_be_important_number(number))
```

```python
def check_seating(n, a):
    occupied = set()
    occupied.add(a[0])

    for i in range(1, n):
        if (a[i] - 1 not in occupied) and (a[i] + 1 not in occupied):
            return "NO"

        occupied.add(a[i])

    return "YES"

def run_tests():
    tests = [
        (5, [5, 4, 2, 1, 3]),  # NO
        (3, [2, 3, 1]),        # YES
        (4, [2, 3, 1, 4]),     # YES
        (5, [1, 2, 3, 5, 4])   # NO
    ]

    expected_results = ["NO", "YES", "YES", "NO"]

    for i, (n, a) in enumerate(tests):
        result = check_seating(n, a)
        assert result == expected_results[i], f"Test {i+1} failed: got {result}, expected {expected_results[i]}"
        print(f"Test {i+1} passed!")

def main():
    t = int(input())
    for _ in range(t):
        n = int(input())
        a = list(map(int, input().split()))
        print(check_seating(n, a))

#run_tests()

if __name__ == "__main__":
    main()
```

```python
def check_pattern(n, a, m, strings):
    results = []
    for s in strings:
        if len(s) != n:
            results.append("NO")
            continue

        num_to_char = {}
        char_to_num = {}
        is_match = True

        for i in range(n):
            num = a[i]
            char = s[i]

            if num in num_to_char:
                if num_to_char[num] != char:
                    is_match = False
                    break
            if char in char_to_num:
                if char_to_num[char] != num:
                    is_match = False
                    break

            num_to_char[num] = char
            char_to_num[char] = num

        if is_match:
            results.append("YES")
        else:
            results.append("NO")

    return results

def run_tests():
    test_cases = [
        # Test case 1
        {
            "n": 5,
            "a": [3, 5, 2, 1, 3],
            "m": 2,
            "strings": ["abfda", "afbfa"],
            "expected": ["YES", "NO"]
        },
        # Test case 2
        {
            "n": 2,
            "a": [1, 2],
            "m": 3,
            "strings": ["ab", "abc", "aa"],
            "expected": ["YES", "NO", "NO"]
        },
        # Test case 3
        {
            "n": 4,
            "a": [5, -3, 5, -3],
            "m": 4,
            "strings": ["aaaa", "bcbc", "aba", "cbcb"],
            "expected": ["NO", "YES", "NO", "YES"]
        },
    ]

    for i, test_case in enumerate(test_cases):
        n = test_case["n"]
        a = test_case["a"]
        m = test_case["m"]
        strings = test_case["strings"]
        expected = test_case["expected"]

        result = check_pattern(n, a, m, strings)
        assert result == expected, f"Test case {i+1} failed: expected {expected}, got {result}"
        print(f"Test case {i+1} passed.")

def main():
    t = int(input())
    all_results = []

    for _ in range(t):
        n = int(input())
        a = list(map(int, input().split()))
        m = int(input())
        strings = [input().strip() for _ in range(m)]

        results = check_pattern(n, a, m, strings)
        all_results.extend(results)

    print("\n".join(all_results))

if __name__ == "__main__":
    #run_tests()
    main()
```

```python
def max_score(n, a, s):
    # Префиксные суммы
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    total_score = 0
    left = 0  # Указатель на L, движущийся слева направо
    right = n - 1  # Указатель на R, движущийся справа налево

    while left < right:
        # Ищем ближайший L слева
        while left < n and s[left] != 'L':
            left += 1
        if left >= right:
            break

        # Ищем ближайший R справа
        while right >= 0 and s[right] != 'R':
            right -= 1
        if left >= right:
            break

        # Если L находится перед R, добавляем сумму и двигаем указатели
        if left < right:
            # Сумма от left до right
            segment_sum = prefix_sum[right + 1] - prefix_sum[left]
            total_score += segment_sum
            # Двигаем указатели дальше
            left += 1
            right -= 1

    return total_score

def run_tests():
    test_cases = [
        (6, [3, 5, 1, 4, 3, 2], "LRLLLR", 18),
        (2, [2, 8], "LR", 10),
        (2, [3, 9], "RL", 0),
        (5, [1, 2, 3, 4, 5], "LRLRR", 22),
    ]

    for i, (n, a, s, expected) in enumerate(test_cases):
        result = max_score(n, a, s)
        assert result == expected, f"Test case {i+1} failed: expected {expected}, got {result}"
        print(f"Test case {i+1} passed")

def main():
    import sys
    input = sys.stdin.read
    data = input().split()

    index = 0
    t = int(data[index])
    index += 1
    results = []

    for _ in range(t):
        n = int(data[index])
        index += 1
        a = list(map(int, data[index:index+n]))
        index += n
        s = data[index]
        index += 1
        results.append(max_score(n, a, s))

    for result in results:
        print(result)

if __name__ == "__main__":
    #run_tests()
    main()
```
